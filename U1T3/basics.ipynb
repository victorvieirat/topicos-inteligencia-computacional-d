Create a function, named knn_train_test() that encapsulates the training and simple validation process. This function should have 3 parameters -- training column name, target column name, and the dataframe object.
This function should split the data set into a training and test set.

```python
from sklearn.neighbors import KNeighborsRegressor
# I will use the Median Absolute Error as metric
from sklearn.metrics import median_absolute_error


def knn_train_test(train_col, target_col, df):
    knn = KNeighborsRegressor()
    np.random.seed(40028922)

    shuffled_index = np.random.permutation(df.index)
    df_random = df.reindex(shuffled_index)

    prev_train = int(len(df_random) / 2)

    df_train = df_random.iloc[0:prev_train]
    df_test = df_random.iloc[prev_train:]

    knn.fit(df_train[[train_col]], df_train[target_col])
    predicted_labels = knn.predict(df_test[[train_col]])

    mae = median_absolute_error(df_test[target_col], predicted_labels)
    rmae = np.sqrt(mae)
    return rmae


# pep8(_ih)
```

```python
rmae_results = {}
train_cols = cars_numeric.columns.drop('price')

for col in train_cols:
    rmae_val = knn_train_test(col, 'price', cars_numeric)
    rmae_results[col] = rmae_val
print('Median Absolute Error')
rmae_results_series = pd.Series(rmae_results)
rmae_results_series.sort_values()
# pep8(_ih)
```

```python
def knn_train_test(train_col, target_col, df):
    np.random.seed(40028922)

    mixed_index = np.random.permutation(df.index)
    df_random = df.reindex(mixed_index)

    prev_train = int(len(df_random) / 2)

    df_train = df_random.iloc[0:prev_train]
    df_test = df_random.iloc[prev_train:]

    k_rmeas = {}

    for k in 1, 3, 5, 7, 9:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(df_train[[train_col]], df_train[target_col])

        predicted_labels = knn.predict(df_test[[train_col]])

        mea = median_absolute_error(df_test[target_col], predicted_labels)
        rmea = np.sqrt(mea)

        k_rmeas[k] = rmea
    return k_rmeas


k_rmea_results = {}

train_cols = cars_numeric.columns.drop('price')
for col in train_cols:
    rmea_val = knn_train_test(col, 'price', cars_numeric)
    k_rmea_results[col] = rmea_val

k_rmea_results
# pep8(_ih)
```

```python
import plotly.graph_objects as go

fig = go.Figure()


for k, v in k_rmea_results.items():
    x = list(v.keys())
    y = list(v.values())
    fig.add_trace(go.Scatter(
        x=x,
        y=y,
        name=k,
        mode='lines+markers'
        ))
fig.update_yaxes(title_text='Median Absolute Error')
fig.update_xaxes(title_text='K value')
fig.show()
# pep8(_ih)
```

# Multivariate Model

```python
feature_avg_rmea = {}
for k, v in k_rmea_results.items():
    avg_rmea = np.mean(list(v.values()))
    feature_avg_rmea[k] = avg_rmea
series_avg_rmea = pd.Series(feature_avg_rmea)
sorted_series_avg_rmea = series_avg_rmea.sort_values()
print(sorted_series_avg_rmea)

sorted_features = sorted_series_avg_rmea.index
# pep8(_ih)
```

```python
def knn_train_test(train_cols, target_col, df):
    np.random.seed(1)

    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    prev_train = int(len(rand_df) / 2)

    df_train = rand_df.iloc[0:prev_train]
    df_test = rand_df.iloc[prev_train:]

    k_values = [5]
    k_rmses = {}

    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(df_train[train_cols], df_train[target_col])

        predicted_labels = knn.predict(df_test[train_cols])

        mse = median_absolute_error(df_test[target_col], predicted_labels)
        rmse = np.sqrt(mse)
        k_rmses[k] = rmse
    return k_rmses


k_rmse_results = {}

for nr_best_feats in range(2, 7):
    k_rmse_results['{} best features'.format(nr_best_feats)] = knn_train_test(
        sorted_features[:nr_best_feats],
        'price',
        cars_numeric
        )

k_rmse_results
# pep8(_ih)
```

# Hyperparameter Tuning

```python
def knn_train_test(train_cols, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    last_train_row = int(len(rand_df) / 2)
    df_train = rand_df.iloc[0:last_train_row]
    df_test = rand_df.iloc[last_train_row:]
    k_values = [i for i in range(1, 25)]
    k_rmeas = {}

    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(df_train[train_cols], df_train[target_col])
        predicted_labels = knn.predict(df_test[train_cols])
        mea = median_absolute_error(df_test[target_col], predicted_labels)
        rmea = np.sqrt(mea)

        k_rmeas[k] = rmea
    return k_rmeas


k_rmea_results = {}

for nr_best_feats in range(2, 6):
    k_rmea_results['{} best features'.format(nr_best_feats)] = knn_train_test(
        sorted_features[:nr_best_feats],
        'price',
        cars_numeric
        )

k_rmea_results

# pep8(_ih)
```
